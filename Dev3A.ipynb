```python
# Up to ₹2,50,000      →  No tax
# ₹2,50,001 – ₹5,00,000 →  5%
# ₹5,00,001 – ₹10,00,000 → 20%
# Above ₹10,00,000      →  30%

# Enter annual income : 750000

# --- Tax Breakdown ---
# Income         : ₹7,50,000
# Tax Slab       : 20%
# Tax Amount     : ₹1,50,000
# Net Income     : ₹6,00,000

# 💚 Hint: Tax = (income - slab_start) × rate. For 750000 → it falls in 20% slab → (750000-500000) × 0.20 = 50000. Think carefully!

```

```python
ic = int(input("Enter annual income : ").strip())

if ic <= 250000:
    rate = 0.0
    slab_start = 0
elif ic <= 500000:
    rate = 0.05
    slab_start = 250000
elif ic <= 1000000:
    rate = 0.20
    slab_start = 500000
else:
    rate = 0.30
    slab_start = 1000000

# Calculate tax
ta = (ic - slab_start) * rate
ni = ic - ta

print("\n--- Tax Breakdown ---")
print(f"Income     : ₹{ic}")
print(f"Tax Slab   : {'No Tax' if rate == 0.0 else f'{int(rate*100)}%'}")
print(f"Tax Amount : ₹{ta:.2f}")
print(f"Net Income : ₹{ni}")
```

```python
n = int(input("Enter the number:").strip())
new_rev = 0
while n > 0:
    last_digit = n % 10
    new_rev = new_rev * 10 + last_digit
    n = n // 10    
print(f"Reverse : {new_rev}")
```
